```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import io
import pandas as pd

data = files.upload()
```

```python
df = pd.read_csv(io.StringIO(data['Housing.csv'].decode('utf-8')))
```

```python

# Sorting
# Filtering (e.g., subset data based on conditions)
# Grouping (e.g., by categorical variables)



# Sort the DataFrame by 'price' in descending order
df_sorted = df.sort_values('price', ascending=False)

# Filter the DataFrame to include only houses with more than 2 bedrooms and price greater than 300000
df_filtered = df[ (df['bedrooms'] > 2) & (df['price'] > 300000) ]

# Group data by 'stories' and calculate the mean price for each group
df_grouped = df.groupby('stories')['price'].mean()

# Print the manipulated dataframes (optional)
print("Sorted Data:")
print(df_sorted.head()) # prints the first few rows
print("\nFiltered Data:")
print(df_filtered.head())
print("\nGrouped Data:")
df_grouped
```

```python


import matplotlib.pyplot as plt

# Box plot
plt.figure(figsize=(8, 6))
df.boxplot(column='price')
plt.title('Box Plot of Housing Prices')
plt.ylabel('Price')
plt.show()

# Pie chart (example using 'bedrooms' column)
bedrooms_counts = df['bedrooms'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(bedrooms_counts, labels=bedrooms_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Bedrooms')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Correlation analysis
print("\nCorrelation Matrix:")
# Select only numerical features for correlation analysis
numerical_df = df.select_dtypes(include=np.number)
print(numerical_df.corr())

# Distribution analysis (using histograms for numerical features)
numerical_features = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    plt.hist(df[feature], bins=20, edgecolor='black')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

# Additional analysis (example: price distribution by number of bedrooms)
plt.figure(figsize=(10, 6))
df.groupby('bedrooms')['price'].plot(kind='kde', legend=True)
plt.title('Price Distribution by Number of Bedrooms')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()
```
